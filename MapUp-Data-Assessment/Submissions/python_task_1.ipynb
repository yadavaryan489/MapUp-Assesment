### Question 2: Car Type Count Calculation

```python
def get_type_count(df):

  # Define car type thresholds
  thresholds = {
      "low": 15,
      "medium": 25
  }

  # Create a new column 'car_type' based on car values
  df["car_type"] = pd.cut(df["car"], bins=[-float("inf")] + list(thresholds.values()) + [float("inf")],
                           labels=list(thresholds.keys()) + ["high"])

  # Calculate the count of occurrences for each car type
  type_counts = df["car_type"].value_counts().to_dict()

  # Sort the dictionary alphabetically based on keys
  sorted_counts = dict(sorted(type_counts.items()))

  return sorted_counts

type_counts = get_type_count(df1)


type_counts
```

### Question 3: Bus Count Index Retrieval

```python

def get_bus_indexes(df):

  # Calculate the mean of the bus column
  mean_bus = df["bus"].mean()

  # Identify bus values greater than twice the mean
  filtered_df = df[df["bus"] > 2 * mean_bus]

  # Extract and sort the bus indices
  bus_indexes = sorted(filtered_df.index.to_list())

  return bus_indexes

# Create a sample DataFrame
# Get the bus indexes
bus_indexes = get_bus_indexes(df1)

# Print the bus indexes
bus_indexes
```

```python
def filter_routes(df):

  # Calculate average truck values per route
  avg_truck_per_route = df.groupby("route")["truck"].mean()

  # Filter routes with average truck values above 7
  filtered_routes = avg_truck_per_route[avg_truck_per_route > 7]

  # Return sorted list of filtered route values
  return sorted(filtered_routes.index.to_list())

# Create a sample DataFrame


# Filter and sort routes based on average truck values
filtered_routes = filter_routes(df1)

# Print the filtered routes
print(filtered_routes)
```

```python
def multiply_matrix(df):

  # Define multipliers for values above and below 20
  multiplier_high = 0.75
  multiplier_low = 1.25

  def apply_multiplier(value, dtype=int):

    try:
      value = dtype(value)
    except ValueError:
      # Handle non-numeric values appropriately
      return value

    if value > 20:
      return value * multiplier_high
    else:
      return value * multiplier_low

  # Apply the function to each value in the DataFrame
  modified_df = df.applymap(apply_multiplier)

  # Round all values to 1 decimal place
  modified_df = modified_df.round(1)

  return modified_df
modified_df = multiply_matrix(car_matrix)
modified_df
```
